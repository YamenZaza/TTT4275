```python
# kNN model prediction
def sk_knn_prediction(model, test_i):
    # Takes in a knn model along with a test set and provides the confusion matrix and error
    # Perform predition on the whole test data set - Takes a long time
    print("Prediciton of the test set using the kNN algorithm has started.")
    sk_knn_pred_start_time = time.time()
    sk_knn_pred = model.predict(test_i)
    sk_knn_pred_total_time = time.time() - sk_knn_pred_start_time
    print("It took", sk_knn_pred_total_time, " seconds!")
    np.savetxt('./temp/sk_knn_pred.out', sk_knn_pred)
    print("The predicted results are saved at './temp/sk_knn_pred.out' using numpy module.")
    # Loading
    # sk_knn_pred = np.loadtxt('./temp/sk_knn_pred.out')
    # Confusion matrix and model accuracy
    # Take y_true, y_pred
    score = accuracy_score(test_l, sk_knn_pred)
    print("The total error rate for the KNN algrothm that used the sklearn package was ", 100-score*100, "%")
    # Testing predition
    print("Test: The predicted number was ", model.predict(test_i[1][:, np.newaxis].transpose())[0], " and the test image was")
    plt.imshow(test_i[1].reshape(28,28))
    return confusion_matrix(test_l, sk_knn_pred, labels = np.arange(10))

# kNN model training
def sk_knn_training(train_i, train_l, neighbors):
    # Takes in the training set along wiht the number of neighbors and outputs a knn model of the data
    # Design a KNN classifier with K=7 using euclidean distance for distance measurment
    # KNN performs better with a lower number of features than a large number of features.
    # Research has shown that in large dimension Euclidean distance is not useful anymore.
    # Very important note: The memory restrictions will not allow this block to run more than once. Even if the model is simply saved then load, only one instant of the model can exist in the memory. Restarting the kernel and clearing out helps
    sk_knn_train = np.split(train_i.ravel(), 60000)
    sk_knn_labels = train_l.ravel()
    print("Training of the kNN algorithm has started.")
    sk_knn_start_time = time.time()
    model = KNeighborsClassifier(n_neighbors= neighbors)
    model.fit(sk_knn_train, sk_knn_labels)
    sk_knn_total_time = time.time() - sk_knn_start_time # Training only
    print("It took", sk_knn_total_time, " seconds to train!")
    # Persistent model
    dump(model, './temp/sk_KNN_model.joblib')
    print("The model is saved at './temp/sk_KNN_model.joblib' using joblib module.")
    # Loading KNN model
    # from joblib import load
    # model = load('./temp/sk_KNN_model.joblib')
    return model

# Kmeans models for every number using mini-batch
def sk_kmeans_mini(train_i, train_l):
    # Performs mini-batch kmeans on the given dataset and outputs a new dataset with centroids as tempelates
    # We need to split the training and testing sets into 10 separate classes based on the labels - not equal splits
    # Then perform an M=64 kmeans clustering on each separate training set
    # Use the clusters centers as templates for the nearest neighbors. Instead of using all the 60.000 samples in the first part, now you use 640 vectors.
    # Use mini batch kmeans clustering since we have a big number of training examples
    # We will not make the code more compact or combine everything into arrays since we have unequal sizes
    # More code comments on the fucntion without mini-batch
    class_zero = train_i[np.where(train_l == 0)[0]]
    class_one = train_i[np.where(train_l == 1)[0]]
    class_two = train_i[np.where(train_l == 2)[0]]
    class_three = train_i[np.where(train_l == 3)[0]]
    class_four = train_i[np.where(train_l == 4)[0]]
    class_five = train_i[np.where(train_l == 5)[0]]
    class_six = train_i[np.where(train_l == 6)[0]]
    class_seven = train_i[np.where(train_l == 7)[0]]
    class_eight = train_i[np.where(train_l == 8)[0]]
    class_nine = train_i[np.where(train_l == 9)[0]]
    print("Preparing datasets using kmeans with mini-batch and 64 clusters.")
    sk_mkmeans_start_time = time.time()
    model_zero = MiniBatchKMeans(n_clusters = 64).fit(class_zero)
    model_one = MiniBatchKMeans(n_clusters = 64).fit(class_one)
    model_two = MiniBatchKMeans(n_clusters = 64).fit(class_two)
    model_three = MiniBatchKMeans(n_clusters = 64).fit(class_three)
    model_four = MiniBatchKMeans(n_clusters = 64).fit(class_four)
    model_five = MiniBatchKMeans(n_clusters = 64).fit(class_five)
    model_six = MiniBatchKMeans(n_clusters = 64).fit(class_six)
    model_seven = MiniBatchKMeans(n_clusters = 64).fit(class_seven)
    model_eight = MiniBatchKMeans(n_clusters = 64).fit(class_eight)
    model_nine = MiniBatchKMeans(n_clusters = 64).fit(class_nine)
    sk_mkmeans_total_time = time.time() - sk_mkmeans_start_time # Very fast
    # concatenate all cluster centers
    one = np.ones((64, 1))
    sk_kmeans_labels = np.concatenate((one*0, one*1, one*2, one*3, one*4, one*5, one*6, one*7, one*8, one*9))
    sk_centroid_tempelates = np.concatenate((model_zero.cluster_centers_, model_one.cluster_centers_, model_two.cluster_centers_,
                              model_three.cluster_centers_, model_four.cluster_centers_, model_five.cluster_centers_,
                              model_six.cluster_centers_, model_seven.cluster_centers_, model_eight.cluster_centers_,
                              model_nine.cluster_centers_))
    sk_kmeans_mini_dataset = [sk_centroid_tempelates, sk_kmeans_labels] # list
    print("It took ", sk_mkmeans_total_time, " seconds!")
    dump(sk_kmeans_mini_dataset, './temp/sk_kmeans_mini_dataset.joblib')
    print("The new kmeans dataset is saved at './temp/sk_kmeans_mini_dataset.joblib' using joblib module.")
    # Loading dataset model
    # sk_kmeans_mini_dataset = load('./temp/sk_kmeans_mini_dataset.joblib')
    return sk_kmeans_mini_dataset

# Kmeans models for every number without using mini-batch
def sk_kmeans(train_i, train_l):
    # Takes the trainig images dataset and their labels and outputs the clustering centroid locations as the new
    # trainig set. We go from 10000 to 640 tempelates.
    # Note: In kmeans, each image will be treated as a 784 dimensional point that can be plotted and in that sense we do perform
    # clustering.
    # Group classe tempelates together from the main training dataset
    class_zero = train_i[np.where(train_l == 0)[0]]
    class_one = train_i[np.where(train_l == 1)[0]]
    class_two = train_i[np.where(train_l == 2)[0]]
    class_three = train_i[np.where(train_l == 3)[0]]
    class_four = train_i[np.where(train_l == 4)[0]]
    class_five = train_i[np.where(train_l == 5)[0]]
    class_six = train_i[np.where(train_l == 6)[0]]
    class_seven = train_i[np.where(train_l == 7)[0]]
    class_eight = train_i[np.where(train_l == 8)[0]]
    class_nine = train_i[np.where(train_l == 9)[0]]
    print("Preparing datasets using kmeans with 64 clusters.")
    # Perform kmeans with 64 clusters on every class
    sk_kmeans_start_time = time.time()
    modelx_zero = KMeans(n_clusters = 64).fit(class_zero)
    modelx_one = KMeans(n_clusters = 64).fit(class_one)
    modelx_two = KMeans(n_clusters = 64).fit(class_two)
    modelx_three = KMeans(n_clusters = 64).fit(class_three)
    modelx_four = KMeans(n_clusters = 64).fit(class_four)
    modelx_five = KMeans(n_clusters = 64).fit(class_five)
    modelx_six = KMeans(n_clusters = 64).fit(class_six)
    modelx_seven = KMeans(n_clusters = 64).fit(class_seven)
    modelx_eight = KMeans(n_clusters = 64).fit(class_eight)
    modelx_nine = KMeans(n_clusters = 64).fit(class_nine)
    sk_kmeans_total_time = time.time() - sk_kmeans_start_time
    # concatenate all cluster centers to form the new training set (640 tempelates)
    one = np.ones((64, 1))
    sk_kmeans_labels = np.concatenate((one*0, one*1, one*2, one*3, one*4, one*5, one*6, one*7, one*8, one*9))
    sk_centroid_tempelates = np.concatenate((modelx_zero.cluster_centers_, modelx_one.cluster_centers_, modelx_two.cluster_centers_,
                              modelx_three.cluster_centers_, modelx_four.cluster_centers_, modelx_five.cluster_centers_,
                              modelx_six.cluster_centers_, modelx_seven.cluster_centers_, modelx_eight.cluster_centers_,
                              modelx_nine.cluster_centers_))
    # We return the new tempelate dataset and its labels
    sk_kmeans_dataset = [sk_centroid_tempelates, sk_kmeans_labels] # list
    print("It took ", sk_kmeans_total_time, " seconds!")
    return sk_kmeans_dataset

# Appropriately read binary files of data sets - The MATLAB code will be duplicated in python
def load_sets(train_images, train_labels, test_images, test_labels):
    # The first four elements are big endian int32 and the rest are big endian uint8
    # Unsigned integer is 'u'
    # Big endian is '>'
    # Endianess indicates which values are stored first in memory - Big: most significant - Little: least significant
    # Use offset (in bytes) to skip header after knowing the first four numbers - int32 uses 4 bytes, so 4*4=16
    # numpy.fromfile(file, dtype=float, count=-1, sep='', offset=0)
    # We will keep the data stored in memory as the size is not huge
    # We will read the rest of the file and then use numpy.reshape to form our matrix
    # Each row represents an image in which 'reshape' can be used to convert the row vector into a 28x28 pixel image
    # This method takes much less compared to 'fread' in MATLAB
    # The content of header for binary image files is magic_num, num_train, row_size, col_size
    # The content of header for binay label is magic_num, num_train
    # Big endian uint8 - 8bits = 1byte
    # To access the same element as the one in MATLAB, subtract (1) from both indices

    # Extract images and reshape
    magic_num, num_train, row_size, col_size = np.fromfile(train_images, dtype = '>i4', count = 4)
    vec_size = row_size * col_size
    train_i = np.fromfile(train_images, dtype = '>u1', count = num_train*vec_size, offset = 16).reshape((num_train, vec_size))

    magic_num, num_test, row_size, col_size = np.fromfile(test_images, dtype = '>i4', count = 4)
    vec_size = row_size * col_size
    test_i = np.fromfile(test_images, dtype = '>u1', count = num_test*vec_size, offset = 16).reshape((num_test, vec_size))

    # Extract labels and reshape
    magic_num, num_train = np.fromfile(train_labels, dtype = '>i4', count = 2)
    train_l = np.fromfile(train_labels, dtype = '>u1', count = num_train, offset = 8).reshape((num_train, 1))

    magic_num, num_test = np.fromfile(test_labels, dtype = '>i4', count = 2)
    test_l = np.fromfile(test_labels, dtype = '>u1', count = num_train, offset = 8).reshape((num_test, 1))

    print("Datasets are loaded!")
    print("Training images have size ", train_i.shape)
    print("Training labels have size ", train_l.shape)
    print("Testing images have size ", test_i.shape)
    print("Testing labels have size ", test_l.shape)
    return train_i, train_l, test_i, test_l

# NN using sklearn
# Takes in the training and testing datasets along with chuncking size (optional) and returns NN [tempelate_index, min_distance]
def sk_nn(test_i, train_i, train_tempelate_split = 1):
    print("Training of the NN algorithm has started.")
    sk_nn_start_time = time.time()
    data_iteration = len(train_i)//train_tempelate_split # To prevent memory issues
    final_result = np.zeros((len(test_i), 2)) # [tempelate_index, min_distance]
    for i in range(data_iteration):
        [temp1, temp2] = skmet.pairwise_distances_argmin_min(test_i, train_i[(i*train_tempelate_split):(train_tempelate_split+i*train_tempelate_split), :])
        temp1 = temp1+i*train_tempelate_split # Account for shift in indices
        temp3 = np.concatenate((temp1[:, np.newaxis], temp2[:, np.newaxis]), axis=1)
        if(i==0):
            final_result = np.concatenate((temp1[:, np.newaxis], temp2[:, np.newaxis]), axis=1)
        # We compare the distances between the previous and new iteration. If the new one is less then the index and distance
        # of the new interation is replaced. This is done element-wise
        final_result = np.where(temp3[:, 1][:, np.newaxis]<final_result[:, 1][:, np.newaxis], temp3, final_result)
    sk_nn_total_time = time.time() - sk_nn_start_time
    print("It took ", sk_nn_total_time, " seconds!")
    # Results are saved for easy reusage
    np.savetxt('./temp/final_result_sk_pdam.out', final_result)
    print("The model is saved at ./temp/final_result_sk_pdam.out' using numpy module.")
    # Load results
    # final_result = np.loadtxt('./temp/final_result_sk_pdam.out')
    return final_result, sk_nn_total_time

# Confusion Matrix for NN using sklearn
def sk_nn_conf_error(sk_NN_result, train_l, testing_l):
    y_pred = np.zeros((len(testing_l),1))
    for i in range(len(y_pred)):
        y_pred[i] = train_l[int(sk_NN_result[i,0])][0]
    sk_conf = skmet.confusion_matrix(testing_l, y_pred)
    sk_conf_df = pd.DataFrame(sk_conf, index = ["Zero", "One", "Two", "Three", "Four", "Five", "Six", "Seven",
                    "Eight", "Nine"], columns = ["Zero", "One", "Two", "Three", "Four", "Five", "Six", "Seven",
                    "Eight", "Nine"])
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    # lucidv01d's asnwer from https://stackoverflow.com/questions/31324218/scikit-learn-how-to-obtain-true-positive-true-negative-false-positive-and-fal
    #---
    TP = np.diag(sk_conf_df)
    #---
    # Total error
    error_t = 100 - (sum(TP)/len(testing_l))*100
    print("Overall error rate for the NN classifier is ", error_t, "%")
    display(sk_conf_df)
    # Correctly classified and missclassified indices. A tuple of arrays (one for each axis) containing the indices
    sk_missclass_i = np.where(testing_l != y_pred)
    sk_corrclass_i = np.where(testing_l == y_pred)
    # Plot some results
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(20,20))
    print("Classifier Output Examples are Below!")
    ax1.imshow(test_i[sk_missclass_i[0][0]].reshape(28,28))
    ax1.set_title("Misclassified Example 1: the predicted number is " + str(int(y_pred[sk_missclass_i[0][0]])), fontsize=14, fontweight='bold')
    ax1.axis('off')
    ax2.imshow(test_i[sk_missclass_i[0][1]][:].reshape(28,28))
    ax2.set_title("Misclassified Example 2: the predicted number is " + str(int(y_pred[sk_missclass_i[0][1]])), fontsize=14, fontweight='bold')
    ax2.axis('off')
    ax3.imshow(test_i[sk_corrclass_i[0][1]].reshape(28,28))
    ax3.set_title("Correctly Classified Example 1: the predicted number is " + str(int(y_pred[sk_corrclass_i[0][1]])), fontsize=14, fontweight='bold')
    ax3.axis('off')
    ax4.imshow(test_i[sk_corrclass_i[0][2]].reshape(28,28))
    ax4.set_title("Correctly Classified Example 2: the predicted number is " + str(int(y_pred[sk_corrclass_i[0][2]])), fontsize=14, fontweight='bold')
    ax4.axis('off')
```
